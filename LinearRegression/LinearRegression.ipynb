```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

#### Formula for calculating m and b in Linear Regression
b = y_bar - m.x(bar)

m = COV(x,y)/var(x)

```python
class MyLG:
    def __init__(self,x,y):
        self.x = x
        self.y  =y
        self.m = 0
        self.b = 0
    def fit(self):
        x_bar = self.x.mean()
        y_bar = self.y.mean()
        num = 0
        den = 0
        for xi,yi in zip(self.x,self.y):
            num += (xi-x_bar)*(yi-y_bar)
            den += np.square(xi-x_bar)
        self.m = num/(den+0.00000001)
        self.b = (y_bar - (self.m*x_bar))
    def predict(self, a):
      return [self.m*i + self.b for i in a]
    def getCoeff(self):
        return (self.m , self.b)
        
```

```python
x = np.linspace(0,50,100)

np.random.seed(42)

noise = np.random.normal(0,5, size=x.shape)

y = 2*x + 15 + noise 
```

```python
plt.scatter(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Scatter plot of x vs y')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
lr = MyLG(x_train, y_train)

lr.fit()

y_pred = lr.predict(x)
```

```python
from sklearn.linear_model import LinearRegression

lin = LinearRegression()

x_train_new = x_train.reshape(-1,1)
lin.fit(x_train_new,y_train)

```

```python
x_new = x.reshape(-1,1)
y_pred2 = lin.predict(x_new)
```

```python
plt.scatter(x_train,y_train, alpha=0.7)
plt.scatter(x_test, y_test, alpha=0.7)
plt.plot(x,y_pred, color='red')
plt.plot(x,y_pred2, color='black', alpha=0.5)
plt.show()
```
